`obtener_datos()`:

1. Utilizo "while True" porque necesitamos repetir la pregunta hasta que el dato sea válido, y no sabemos de antemano cuántas veces el usuario se equivocará o si el usuario lo escribirá bien a la primera, a la tercera o a la décima.
2. El bucle infinito + break es la forma de decirle a Python: "no avances hasta que esto esté bien". Hay uno por cada campo para que, si nombre falla, solo se repite la pregunta del nombre, no la del apellido ni la del teléfono. Cada campo es independiente.
3. Cada while por campo manda llamar la función validar_datos() para comprobar que todo esté bien escrito, si la función devuelve un "False", se imprime un mensaje de "[Campo] no válido, inténtalo de nuevo".
4. En el caso del while de experiencia hago un preprocesado: reemplazo comas por puntos, si las hay, y se hace un "casteo" para pasar el dato a número flotante.
5. Para experiencia, catalan, excel y titulación utilizo un try/except para evitar que falle si el usuario escribe letras en vez de números. El pass simplemente deja que el while True vuelva a empezar y pida el dato de nuevo.
6. Al final, la función retorna un diccionario por candidato.


```python
def obtener_datos():

    while True:
        nombre = input("Ingresa tu nombre: ")
        if validar_datos("nombre", nombre):
            break
        print("Nombre no válido, inténtalo de nuevo.")

    while True:
        apellido = input("Ingresa tu apellido: ")
        if validar_datos("apellido", apellido):
            break
        print("Apellido no válido, inténtalo de nuevo.")

    while True:
        telefono = input("Ingresa tu número de teléfono: ")
        if validar_datos("telefono", telefono):
            break
        print("Número no válido, inténtalo de nuevo.")

    while True:
        experiencia = input("Ingresa tus años de experiencia: ")
        experiencia = experiencia.replace(",", ".")
        try:
            experiencia = float(experiencia)
            if validar_datos("experiencia", experiencia):
                break
        except ValueError:
            pass
        print("Años de experiencia no válidos, inténtalo de nuevo.")

    while True:
        catalan = input(
            "Nivel de catalán (introduce un número):\n"
            "  0 - Sin conocimientos\n"
            "  1 - Hasta nivel B1\n"
            "  2 - Nivel B2\n"
            "  3 - C1 o superior\n"
            "Tu respuesta: "
        )
        try:
            catalan = int(catalan)
            if validar_datos("catalan", catalan):
                break
        except ValueError:
            pass
        print("Opción no válida, introduce un número del 0 al 3.")

    while True:
        excel = input(
            "Nivel de Excel (introduce un número):\n"
            "  0 - Nivel básico\n"
            "  1 - Uso de funciones\n"
            "  2 - Tablas dinámicas\n"
            "  3 - Visual Basic y Macros\n"
            "Tu respuesta: "
        )
        try:
            excel = int(excel)
            if validar_datos("excel", excel):
                break
        except ValueError:
            pass
        print("Opción no válida, introduce un número del 0 al 3.")

    while True:
        titulacion = input(
            "Nivel de estudios (introduce un número):\n"
            "  0 - ESO\n"
            "  1 - FP\n"
            "  2 - Universidad\n"
            "  3 - Máster o superior\n"
            "Tu respuesta: "
        )
        try:
            titulacion = int(titulacion)
            if validar_datos("titulacion", titulacion):
                break
        except ValueError:
            pass
        print("Opción no válida, introduce un número del 0 al 3.")

    return {
        "nombre": nombre,
        "apellido": apellido,
        "telefono": telefono,
        "experiencia": experiencia,
        "catalan": catalan,
        "excel": excel,
        "titulacion": titulacion
    }
```

`validar_datos()`:
1. La función recibe un campo y un valor a la vez.
2. Si el campo es nombre o apellido, elimina los espacios y comprueba que todos los caracteres son letras. Devuelve "True" si es así.
3. Si el campo es el teléfono, también se eliminan los espacios y devuelve "True" si los caracteres son dígitos y la longitud es menor o igual a 11.
4. Si el campo es experiencia, el valor debe estar entre 0 y 100.
5. Si se trata de los campos catalán, excel o titulación, se asegura de que el valor sea entre 0 y 3.
6. Si ninguno de estos if y elif funciona, se retorna un "False" y obtener_datos() muestra un mensaje de "vuelve a intentarlo" hasta que se ingrese el dato correcto.


```python
def validar_datos(campo, valor):
    if campo == "nombre" or campo == "apellido":
        return valor.replace(" ", "").isalpha()
    elif campo == "telefono":
        telefono_limpio = valor.replace(" ", "")
        return telefono_limpio.isdigit() and len(telefono_limpio) <= 11
    elif campo == "experiencia":
        return 0 <= valor < 100
    elif campo == "catalan" or campo == "excel" or campo == "titulacion":
        return 0 <= valor <= 3
    else:
        return False
```

`clasificar_candidato()`:
1. Recibe el diccionario de un candidato con todos sus datos validados.
2. Calcula la puntuación total aplicando la fórmula:
   (experiencia × 1.5) + catalán + excel + titulación.
3. Clasifica al candidato según su puntuación:
   - ≥ 12 - "Prioridad Alta"
   - entre 8 y 12 - "Candidato Adecuado"
   - < 8 - "No Prioritario"
4. Añade `puntuacion` y `clasificacion` como nuevas claves al diccionario.
5. Devuelve el diccionario actualizado.

```python
def clasificar_candidato(candidato):
    puntuacion_total = (candidato["experiencia"] * 1.5) + candidato["catalan"] + candidato["excel"] + candidato["titulacion"]

    if puntuacion_total >= 12:
        clasificacion = "Prioridad Alta"
    elif 8 <= puntuacion_total < 12:
        clasificacion = "Candidato Adecuado"
    else:
        clasificacion = "No Prioritario"

    candidato["puntuacion"] = puntuacion_total
    candidato["clasificacion"] = clasificacion

    return candidato
```

`mostrar_individual()`:
1. Recibe el diccionario del candidato recién registrado y el contador actual.
2. Imprime nombre, puntuación y clasificación del candidato.
3. Calcula y muestra los lugares restantes disponibles (10 - contador).

```python
def mostrar_individual(candidato, contador):
    print(f"✅ Candidatura registrada!")
    print(f"Nombre: {candidato['nombre']} {candidato['apellido']}")
    print(f"Puntuación: {candidato['puntuacion']}")
    print(f"Clasificación: {candidato['clasificacion']}")
    print(f"Lugares restantes: {10 - contador}")
```

`mostrar_informe()`:
1. Recibe la lista completa de los 10 candidatos registrados.
2. Define un diccionario de orden para traducir la clasificación a un número
   (Prioridad Alta - 1, Candidato Adecuado - 2, No Prioritario - 3).
3. Ordena la lista con sorted() usando ese número como criterio principal
   y la puntuación de mayor a menor como criterio de desempate.
4. Recorre la lista ordenada con un `for` e imprime nombre, teléfono,
   puntuación y clasificación de cada candidato.

```python
def mostrar_informe(candidatos):
    orden = {
        "Prioridad Alta": 1,
        "Candidato Adecuado": 2,
        "No Prioritario": 3
    }

    candidatos_ordenados = sorted(candidatos, key=lambda c: (orden[c["clasificacion"]], -c["puntuacion"]))

    print("\n===== INFORME FINAL DE CANDIDATOS =====")
    for c in candidatos_ordenados:
        print(f"Nombre: {c['nombre']} {c['apellido']}")
        print(f"Teléfono: {c['telefono']}")
        print(f"Puntuación: {c['puntuacion']}")
        print(f"Clasificación: {c['clasificacion']}")
        print("---")
```

`main_seleccion()`:
1. Es la función principal que orquesta todo el programa.
2. Inicializa la lista vacía candidatos y el contador en 0.
3. Ejecuta un bucle `while contador < 10` que se repite hasta registrar
   exactamente 10 candidatos. En cada iteración:
   - Llama a obtener_datos() y guarda el resultado en candidato.
   - Llama a clasificar_candidato() con ese candidato y guarda
     el resultado actualizado.
   - Añade el candidato a la lista con append().
   - Llama a mostrar_individual() para mostrar el resultado.
   - Suma 1 al contador con contador += 1.
4. Cuando el contador llega a 10, sale del bucle y llama a
   mostrar_informe() con la lista completa.


```python
def main_seleccion():
    candidatos = []
    contador = 0

    while contador < 10:
        print(f"\n--- Candidato {contador + 1} de 10 ---")
        candidato = obtener_datos()
        candidato = clasificar_candidato(candidato)
        candidatos.append(candidato)
        mostrar_individual(candidato, contador + 1)
        contador += 1

    mostrar_informe(candidatos)
```

```python
import pandas as pd
```

`obtener_factor_edad()`:
1. La función recibe los rangos de edad de los clientes. Como si fuera una "columna" de la tabla clientes.
2. Se crea un diccionario de lookup (como una tabla de consulta) para asignar un factor (número decimal) a cada rango de edad sin necesidad de if/elif.
3. Cada grupo de edad de los clientes se encuentra en el diccionario factores_edad y se asigna un número decimal al que llamamos "factor" y guardamos en esa variable.
4. La función devuelve el factor.

```python
def obtener_factor_edad(grupo_edad):
  factores_edad = { "18-25": 1.20, "26-40": 1.00, "41-60": 0.85, "60+": 0.70 }
  factor = factores_edad[grupo_edad]
  return factor
```

`obtener_segmento_y_factor()`:
1. Similar a la función anterior, se recibe como parámetro el código de comportamiento de los clientes.
2. Se crean dos diccionarios de lookup, uno para los factores de comportamiento y otro para los segmentos.
3. La función devuelve un código de comportamiento (float) y un segmento (string) por cada cliente de la lista.

```python
def obtener_segmento_y_factor(codigo_comportamiento):
    factores_comportamiento = {0: 0.10, 1: 0.25, 2: 0.50, 3: 0.75, 4: 0.90}
    segmentos = {0: "Frío", 1: "Frío", 2: "Tibio", 3: "Caliente", 4: "Caliente"}
    return factores_comportamiento[codigo_comportamiento], segmentos[codigo_comportamiento]
```

`calcular_riesgo()`:
1. La función recibe los factores antes calculados (float).
2. Se realiza una operación matemática simple: multiplicación de los factores.
3. La función devuelve el resultado de la multiplicación, llamándole "riesgo".

```python
def calcular_riesgo(factor_codigo, factor_edad):
    riesgo = factor_codigo * factor_edad
    return riesgo
```

`procesar_clientes()`:
1. La función recibe toda la lista que contiene los datos de los clientes en diccionarios.
2. Se define el diccionario de mensajes para que se asignen según el segmento al que pertenece el cliente.
4. Con un bucle for que recorre la lista de clientes, la actualizamos añadiendo las "nuevas columnas" calculadas (factor_edad, factor_codigo, segmento, riesgo_final y mensaje).
5. La función devuelve la lista de clientes actualizada

```python
def procesar_clientes(clientes):
    mensajes = {
        "Frío":     "Descubre nuestros productos destacados",
        "Tibio":    "Tienes artículos esperándote, vuelve a verlos",
        "Caliente": "Tu cesta te espera, completa la compra ahora"
    }
    for cliente in clientes:
        cliente["factor_edad"]   = obtener_factor_edad(cliente["grupo_edad"])
        factor_codigo, segmento  = obtener_segmento_y_factor(cliente["codigo_comportamiento"])
        cliente["factor_codigo"] = factor_codigo
        cliente["segmento"]      = segmento
        cliente["riesgo_final"]  = calcular_riesgo(factor_codigo, cliente["factor_edad"])
        cliente["mensaje"]       = mensajes[segmento]
    return clientes
```

`mostrar_dataframe()`:
1. La función recibe la lista actualizada de clientes como parámetro.
2. La lista se convierte en un dataframe con la librería de pandas.
3. Utilizando funciones de pandas, con sort_values() se reordenan las filas del DataFrame por segmento y riesgo para que en la salida aparezcan primero todos los "Caliente", luego los "Tibio" y finalmente los "Frío", cada grupo ordenado por riesgo de mayor a menor.
4. Se imprime la línea "DATOS DE CLIENTES" y to_string(index=False) imprime el DataFrame sin la columna de índices numéricos que pandas añade por defecto, para que quede más limpio.

```python
def mostrar_dataframe(clientes):
    df = pd.DataFrame(clientes)
    df = df.sort_values(by=["segmento", "riesgo_final"], ascending=[True, False])
    print("\n===== DATOS DE CLIENTES =====")
    print(df.to_string(index=False))
```

5. Pregunta 2: codigos es un `set` que extrae los valores únicos de

6. Un ciclo for recorre los códigos ordenados con `sorted()` de menor a mayor.
Para cada código, grupo contiene la lista de clientes que comparten ese código.

7. Si `len(grupo) > 1` (más de un cliente con ese código), tiene sentido comparar.
Se imprime el encabezado con el código y el segmento usando `grupo[0]['segmento']`,
ya que todos los clientes del grupo comparten el mismo segmento. Un segundo for
anidado imprime los datos individuales de cada cliente para poder comparar
sus riesgos según grupo de edad.

8. Pregunta 3: frios filtra los clientes cuyo segmento es "Frío" y un for
los imprime uno por uno.

9. Pregunta 4: `max()` busca el cliente con el `riesgo_final` más alto dentro
de `calientes`, no de todos los clientes, ya que el más rentable debe ser
alguien que ya estaba cerca de comprar. Finalmente se imprimen sus datos.

```python
def mostrar_respuestas(clientes):

    # Pregunta 1: ¿Qué clientes contactar primero?
    print("\n===== 1. CLIENTES A CONTACTAR PRIMERO =====")
    calientes = [c for c in clientes if c["segmento"] == "Caliente"]
    calientes_ordenados = sorted(calientes, key=lambda c: c["riesgo_final"], reverse=True)
    for c in calientes_ordenados:
        print(f"  {c['nombre']} - Riesgo: {c['riesgo_final']:.2f} - Mensaje: {c['mensaje']}")

    # Pregunta 2: ¿Diferencias entre mismo comportamiento y distinta edad?
    print("\n===== 2. COMPARATIVA POR COMPORTAMIENTO Y EDAD =====")
    codigos = set(c["codigo_comportamiento"] for c in clientes)
    for codigo in sorted(codigos):
        grupo = [c for c in clientes if c["codigo_comportamiento"] == codigo]
        if len(grupo) > 1:
            print(f"\n  Código {codigo} — Segmento: {grupo[0]['segmento']}:")
            for c in grupo:
                print(f"    {c['nombre']} | Edad: {c['edad']} | Grupo: {c['grupo_edad']} | Riesgo: {c['riesgo_final']:.2f}")

    # Pregunta 3: ¿Qué clientes no vale la pena contactar?
    print("\n===== 3. CLIENTES DE BAJO INTERÉS =====")
    frios = [c for c in clientes if c["segmento"] == "Frío"]
    for c in frios:
        print(f"  {c['nombre']} - Riesgo: {c['riesgo_final']:.2f} - Segmento: {c['segmento']}")

    # Pregunta 4: ¿Qué cliente es más rentable recuperar?
    print("\n===== 4. CLIENTE MÁS RENTABLE =====")
    mas_rentable = max(calientes, key=lambda c: c["riesgo_final"])
    print(f"  {mas_rentable['nombre']} - Riesgo: {mas_rentable['riesgo_final']:.2f} - Mensaje: {mas_rentable['mensaje']}")
```

# NIVEL 2 / Ejercicio 1:  Gestión de riesgos y priorización de pacientes

## Resultado


🧠 **Algoritmo inicial**

1. Recibir y transformar los registros de los pacientes en crudo.
2. Se corrige la descripción de los síntomas.
3. Se calcula el nivel de riesgo según los síntomas.
4. Se procesa a todos los pacientes.
5. Se ordenan los pacientes por prioridad de riesgo y se genera un dataframe.

**Primer desgranaje**

1. Se recibe la lista en crudo con los datos de los pacientes y se corrige (se separan el nombre, edad y síntomas).
2. Se reciben específicamente los síntomas en crudo y se corrigen también (en minúsculas, sin espacios extra, etc).
3. Se recibe la lista del paciente y un diccionario de riesgos. Se suman los niveles de riesgo.
4. La función principal llama a las anteriores para cada paciente y construye la lista de diccionarios final.
5. Se ordena la lista por riesgo de mayor a menor y se imprime el dataframe generado.



**Algoritmo principal**

1. `parsear_paciente(raw_string)`:
Convierte un string crudo "Nombre|edad|síntomas" en un diccionario con las tres partes separadas. Solo hace el split por |.
2. `normalizar_sintomas(sintomas_raw)`:
Recibe el string de síntomas y devuelve una lista limpia: en minúsculas, reemplaza guiones bajos para unificar formatos, elimina espacios extras por cada elemento tras la separación por comas.
3. `calcular_riesgo(sintomas, nivells_risc)`:
Recibe la lista ya normalizada y el diccionario de riesgos. Suma los valores encontrados (los síntomas no reconocidos suman 0 con .get()).
4. `procesar_pacientes(pacients, nivells_risc)`:
La función principal. Llama a las anteriores para cada paciente y construye la lista de diccionarios final.
5. `ordenar_por_prioridad(pacientes_procesados)`:
Ordena la lista por riesgo_total de mayor a menor con sorted() y key=.



```python
import pandas as pd
```

```python
pacients = [
    "Maria|45|tos, Fiebre Alta, dolor_pecho",
    "Luis|33|dolor_cabeza,faTiga",
    "Sara|67|fiebre alta, dificultad RESPIRAR, tos",
    "Jordi|52|fatiga , tos",
    "Anna|29|fiebre alta,dolor_cabeza"
]

nivells_risc = {
    "fiebre_alta": 3,
    "dificultad_respirar": 5,
    "dolor_pecho": 4,
    "tos": 1,
    "fatiga": 1,
    "dolor_cabeza": 1
}
```

`parsear_paciente()`:
1. Divide 'Nombre|edad|síntomas' en sus tres partes.


```python
def parsear_paciente(raw_string):
    nombre, edad, sintomas_raw = raw_string.split("|")
    return {
        "nombre": nombre.strip(),
        "edad": int(edad.strip()),
        "sintomas_raw": sintomas_raw
    }
```

`normalizar_sintomas()`:
1.   Convierte el string de síntomas en una lista limpia:
    minúsculas, sin espacios extra, guiones bajos → espacios.


```python
def normalizar_sintomas(sintomas_raw):
    return [
        s.strip().lower().replace(" ", "_")
        for s in sintomas_raw.split(",")
        if s.strip()  # descarta tokens vacíos
    ]
```

`calcular_riesgo()`:
1. Suma el valor de riesgo de cada síntoma reconocido.


```python
def calcular_riesgo(sintomas, nivells_risc):
    return sum(nivells_risc.get(s, 0) for s in sintomas)
```

`procesar_pacientes()`:
1. Orquesta el pipeline completo y devuelve lista de dicts.


```python
def procesar_pacientes(pacients, nivells_risc):
    resultado = []
    for raw in pacients:
        paciente = parsear_paciente(raw)
        sintomas = normalizar_sintomas(paciente["sintomas_raw"])
        riesgo   = calcular_riesgo(sintomas, nivells_risc)
        resultado.append({
            "nombre":            paciente["nombre"],
            "edad":              paciente["edad"],
            "sintomas":          sintomas,
            "riesgo_total":      riesgo
        })
    return resultado
```

`ordenar_por_prioridad()`:
1. Devuelve la lista ordenada de mayor a menor riesgo.


```python
def ordenar_por_prioridad(pacientes):
    return sorted(pacientes, key=lambda p: p["riesgo_total"], reverse=True)
```

```python
pacientes_procesados = procesar_pacientes(pacients, nivells_risc)
pacientes_ordenados  = ordenar_por_prioridad(pacientes_procesados)

df = pd.DataFrame(pacientes_ordenados)
print(df.to_string(index=False))
```
